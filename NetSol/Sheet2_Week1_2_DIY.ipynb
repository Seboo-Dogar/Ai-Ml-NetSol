
# Exercise 1: Basic Vector Operations

This guide explains common vector operations in linear algebra using NumPy arrays. Let’s take two 3D vectors as examples:

v1 = [1, 2, 3]
v2 = [4, 5, 6]


---

## 1. ➕ Vector Addition

**Concept**: Add corresponding elements of the two vectors.

**Example**:

v1 + v2 = [1+4, 2+5, 3+6] = [5, 7, 9]


---

## 2. ➖ Vector Subtraction

**Concept**: Subtract corresponding elements of one vector from another.

**Example**:

v1 - v2 = [1−4, 2−5, 3−6] = [−3, −3, −3]


---

## 3. 🔸 Dot Product

**Concept**: Multiply corresponding elements and sum them.

**Formula**:
dot(v1, v2) = 1×4 + 2×5 + 3×6 = 4 + 10 + 18 = 32


**Result**: A scalar value.

---

## 4. ✖️ Cross Product

**Concept**: The result is a vector perpendicular to both input vectors (in 3D space).

**Formula** (for 3D vectors):

v1 × v2 = [
(2×6 − 3×5), # i
(3×4 − 1×6), # j
(1×5 − 2×4) # k
] = [-3, 6, -3]


---

## 5. 📏 Vector Magnitude

**Concept**: The magnitude (or length) of a vector is calculated using the square root of the sum of the squares of its components.

**Formula**:
|v1| = √(1² + 2² + 3²) = √(1 + 4 + 9) = √14 ≈ 3.74



```python
import numpy as np

# Create two vectors
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# 1. Vector addition
def vector_addition(a, b):
    # Your code here
    return a + b

# 2. Vector subtraction
def vector_subtraction(a, b):
    # Your code here
    return a - b

# 3. Dot product
def dot_product(a, b):
    # Your code here
    return np.dot(a, b)

# 4. Cross product (for 3D vectors)
def cross_product(a, b):
    # Your code here
    return np.cross(a, b)

# 5. Vector magnitude
def vector_magnitude(a):
    # Your code here
    return np.linalg.norm(a)

# Test your functions
print("Addition:", vector_addition(v1, v2))
print("Subtraction:", vector_subtraction(v1, v2))
print("Dot product:", dot_product(v1, v2))
print("Cross product:", cross_product(v1, v2))
print("Magnitude v1:", vector_magnitude(v1))
```

# Exercise 2: Basic Matrix Operations

This section explains common matrix operations in linear algebra using NumPy arrays. Let’s take two 2×2 matrices as examples:

m1 = [[1, 2],
[3, 4]]

m2 = [[5, 6],
[7, 8]]


---

## 1. ➕ Matrix Addition

**Concept**: Add corresponding elements of the two matrices.

**Example**:
m1 + m2 = [[1+5, 2+6],
[3+7, 4+8]]
= [[6, 8],
[10, 12]]


---

## 2. ➖ Matrix Subtraction

**Concept**: Subtract corresponding elements of one matrix from another.

**Example**:
m1 - m2 = [[1−5, 2−6],
[3−7, 4−8]]
= [[-4, -4],
[-4, -4]]


---

## 3. ✖️ Matrix Multiplication

**Concept**: Multiply rows of the first matrix with columns of the second matrix (not element-wise).

**Example**:
m1 × m2 = [[(1×5 + 2×7), (1×6 + 2×8)],
[(3×5 + 4×7), (3×6 + 4×8)]]
= [[19, 22],
[43, 50]]


---

## 4. 🔁 Matrix Transpose

**Concept**: Flip the matrix over its diagonal (swap rows and columns).

---

## 5. 📐 Matrix Determinant (for square matrices)

**Concept**: A scalar value that represents a matrix's scaling factor; used to determine invertibility.

**Formula for 2×2**:
det(m1) = (1×4) − (2×3) = 4 − 6 = -2








```python
# Create two matrices
m1 = np.array([
    [1, 2],
    [3, 4]
])

m2 = np.array([
    [5, 6],
    [7, 8]
])

# 1. Matrix addition
def matrix_addition(a, b):
    # Your code here
    return a + b

# 2. Matrix subtraction
def matrix_subtraction(a, b):
    # Your code here
    return a - b

# 3. Matrix multiplication
def matrix_multiplication(a, b):
    # Your code here
    return np.matmul(a, b)

# 4. Matrix transpose
def matrix_transpose(a):
    # Your code here
    return np.transpose(a)

# 5. Matrix determinant (for square matrices)
def matrix_determinant(a):
    # Your code here
    return np.linalg.det(a)

# Test your functions
print("Matrix addition:\n", matrix_addition(m1, m2))
print("Matrix subtraction:\n", matrix_subtraction(m1, m2))
print("Matrix multiplication:\n", matrix_multiplication(m1, m2))
print("Matrix transpose m1:\n", matrix_transpose(m1))
print("Matrix determinant m1:", matrix_determinant(m1))
```

# Exercise 3: Vector Transformations and Angle Calculation

This section covers fundamental vector transformations and angle calculations in linear algebra. We'll use these sample vectors:

v1 = [1, 2, 3]  
v2 = [4, 5, 6]  
v = [1, 0]  # Special 2D vector for rotation example

---

## 1. ⚖️ Vector Scaling

**Concept**: Multiply a vector by a scalar to change its magnitude without changing its direction (unless scalar is negative).

**Formula**:  
scaled_v = scalar × v

**Example**:  
Scale v1 by 2.5:  
2.5 × [1, 2, 3] = [2.5, 5.0, 7.5]

**Visualization**:  
Original: → (length 1)  
Scaled by 2.5: →––––– (length 2.5)

---

## 2. 📐 Angle Between Vectors

**Concept**: Calculate the angle θ between two vectors using dot product.

**Formula**:  
θ = arccos((v1·v2) / (||v1|| × ||v2||))

**Calculation Steps**:
1. Compute dot product: 1×4 + 2×5 + 3×6 = 32
2. Compute magnitudes: ||v1|| ≈ 3.74, ||v2|| ≈ 8.77
3. cosθ = 32 / (3.74 × 8.77) ≈ 0.974
4. θ ≈ arccos(0.974) ≈ 0.23 radians (≈13.2°)

**Special Cases**:
- 0°: Parallel vectors ([1,0] vs [2,0])
- 90°: Perpendicular vectors ([1,0] vs [0,1])

---

## 3. ➡️ Vector Projection

**Concept**: Projection of vector a onto b shows how much of a points in b's direction.

**Formula**:  
proj_b a = ((a·b)/(b·b)) × b

**Example**:  
Project v1 onto v2:
1. Dot product (a·b) = 32 (from previous)
2. b·b = 4² + 5² + 6² = 77
3. Projection = (32/77) × [4,5,6] ≈ [1.66, 2.08, 2.49]

**Visualization**:  
Imagine v1's "shadow" cast onto v2's direction.

---

## 4. 🔄 2D Vector Rotation

**Concept**: Rotate a vector by angle θ in 2D space using rotation matrix.

**Rotation Matrix**:  
R = [[cosθ, -sinθ],  
     [sinθ,  cosθ]]

**Formula**:  
v_rotated = R × v

**Example**:  
Rotate v = [1,0] by 90° (π/2 radians):
1. cos(90°) = 0, sin(90°) = 1
2. Calculation:
   [0 -1] × [1] = [0×1 + -1×0] = [0]  
   [1  0]   [0]   [1×1 +  0×0]   [1]
3. Result: [0,1] (points straight up)

**Common Rotations**:
- 90°: [x,y] → [-y,x]
- 180°: [x,y] → [-x,-y]
- 270°: [x,y] → [y,-x]



```python
# 1. Vector scaling
def scale_vector(v, scalar):
    # Your code here
    return v * scalar

# 2. Angle between two vectors (in radians)
def angle_between_vectors(a, b):
    # Your code here
    dot = np.dot(a, b)
    norms = np.linalg.norm(a) * np.linalg.norm(b)
    cos_theta = dot / norms
    return np.arccos(np.clip(cos_theta, -1.0, 1.0))

# 3. Vector projection of a onto b
def vector_projection(a, b):
    # Your code here
    b_norm = np.linalg.norm(b)
    if b_norm == 0:
        return np.zeros_like(a)
    return (np.dot(a, b) / b_norm**2) * b

# 4. Rotate a 2D vector by theta radians
def rotate_vector_2d(v, theta):
    # Your code here
    cos_theta = np.cos(theta)
    sin_theta = np.sin(theta)
    rotation_matrix = np.array([[cos_theta, -sin_theta], [sin_theta, cos_theta]])
    return np.matmul(rotation_matrix, v)

# Test your functions
v = np.array([1, 0])
print("Scaled vector:", scale_vector(v, 2.5))
print("Angle between v1 and v2:", angle_between_vectors(v1, v2))
print("Projection of v1 onto v2:", vector_projection(v1, v2))
print("Rotated vector:", rotate_vector_2d(v, np.pi/2))  # 90 degrees
```

# Exercise 4: Eigenvalues and Eigenvectors

## 1. Core Concept
Eigenvalues (λ) and eigenvectors (v) are fundamental concepts in linear algebra that satisfy the equation:

A × v = λ × v

where:
- **A** is a square matrix
- **v** is a non-zero eigenvector
- **λ** is a scalar eigenvalue

## 2. Key Properties
- **Invariant Direction**: Eigenvectors point in directions unchanged by the transformation
- **Scaling Factor**: Eigenvalues show how much eigenvectors are scaled
- **Matrix Characterization**: Reveal fundamental properties of the matrix

## 3. Computation Steps

### Example Matrix:

A = [[ 2, -1],
[-1, 2]]


### Step 1: Find Eigenvalues
Solve the characteristic equation:
det(A - λI) = 0


**Calculation**:
1. Subtract λ from diagonal:
[[2-λ, -1],
[ -1, 2-λ]]
2. Compute determinant:

(2-λ)(2-λ) - (-1)(-1) = λ² - 4λ + 3 = 0

3. Solve quadratic equation:
λ₁ = 3, λ₂ = 1


### Step 2: Find Eigenvectors
For each λ, solve (A - λI)v = 0

**For λ₁ = 3**:

[[-1, -1], × [x] = [0]
[-1, -1]] [y] [0]

Solution: v₁ = [1, -1] (or any multiple)

**For λ₂ = 1**:

[[1, -1], × [x] = [0]
[-1, 1]] [y] [0]

Solution: v₂ = [1, 1] (or any multiple)

## 4. Verification
Check A × v = λ × v:

**First eigenpair**:

[[2,-1], × [ 1] = [3] = 3 × [ 1]
[-1,2]] [-1] [-3] [-1]

```python
# 1. Compute eigenvalues and eigenvectors
def compute_eigen(matrix):
    # Your code here
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    return eigenvalues, eigenvectors

# 2. Verify that Av = λv for the first eigenvalue/eigenvector pair
def verify_eigen(matrix, eigenvalues, eigenvectors):
    # Your code here
    v = eigenvectors[:, 0]
    # λ
    lambda_ = eigenvalues[0] 

    Av = np.matmul(matrix, v)
    # λv
    lambda_v = lambda_ * v  

    # Check if they are approximately equal
    if np.allclose(Av, lambda_v):
        print("Verification successful: Av is approximately equal to λv")
        print("Av:", Av)
        print("λv:", lambda_v)
    else:
        print("Verification failed: Av is not approximately equal to λv")
        print("Av:", Av)
        print("λv:", lambda_v)

# Test with a sample matrix
sample_matrix = np.array([[2, -1], [-1, 2]])
eigenvalues, eigenvectors = compute_eigen(sample_matrix)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
verify_eigen(sample_matrix, eigenvalues, eigenvectors)
```

# Exercise 5: Grouping Data Points by Angular Similarity

**Angle Calculation:**

Convert vectors to unit vectors first

Use dot product: cosθ = (a·b)/(||a||·||b||)

Remember: arccos returns radians (convert to degrees)


**Grouping Logic:**

Points belong to same group if angle between them < threshold

Start new group when no similar points found

Simple iterative approach


**Visual Interpretation:**

Points with similar direction get same color

Threshold controls group granularity

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Using only sepal length and width
y = iris.target

def calculate_angle(a, b):
    """Calculate the angle between two vectors in degrees"""
    # Normalize vectors a and b (convert to unit vectors)
    unit_a = a / np.linalg.norm(a)
    unit_b = b / np.linalg.norm(b)
    
    # Compute dot product
    dot_product = np.dot(unit_a, unit_b)
    
    # Use arccos to get angle in radians
    angle_rad = np.arccos(np.clip(dot_product, -1, 1))  # clip to avoid numerical issues
    
    # Convert to degrees
    angle_deg = np.degrees(angle_rad)
    
    return angle_deg

def group_by_angle(data, angle_threshold=30):
    """Group data points based on angular similarity"""
    n_samples = data.shape[0]
    groups = np.zeros(n_samples, dtype=int)
    current_group = 0

    for i in range(n_samples):
        if groups[i] == 0:  # If point isn't assigned yet
            groups[i] = current_group
            for j in range(i+1, n_samples):
                angle = calculate_angle(data[i], data[j])
                # If angle < threshold, assign to same group
                if angle < angle_threshold:
                    groups[j] = current_group
            current_group += 1
    return groups

# Run grouping
angle_threshold = 30  # degrees
groups = group_by_angle(X, angle_threshold)

# Visualize results
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=groups, cmap='tab20', alpha=0.7)
plt.title(f"Data Grouping by Angle (Threshold: {angle_threshold}°)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.colorbar(label='Group ID')
plt.grid(True)
plt.show()
```
